```python
import numpy as np
import matplotlib.pyplot as plt

# Функция для оценки параметра λ методом максимального правдоподобия
def estimate_lambda(samples):
    return 1 / np.mean(samples)

# Настоящее значение параметра λ
true_lambda = 7

# Объемы выборок
sample_sizes = [10, 50, 100, 500, 1000]

# Число повторов для каждой выборки
num_repeats = 1000
    
# Для каждого объема выборки будем хранить оценки λ
estimates = {size: [] for size in sample_sizes}

# Генерация выборок из экспоненциалного распределения для разных размеров выборок и оценка параметра λ
np.random.seed(42)
for size in sample_sizes:
    for _ in range(num_repeats):
        samples = np.random.exponential(1/true_lambda, size)
        estimates[size].append(estimate_lambda(samples))

# Построение графиков
plt.figure(figsize=(14, 8))
for size in sample_sizes:   
    plt.plot(estimates[size], label=f'Объем выборки = {size}')

plt.axhline(y=true_lambda, color='r', linestyle='--', label='Истинное значение λ')
plt.xlabel('Испытание')
plt.ylabel('Оценка λ')
plt.title('Оценка λ для разного объема выборки')
plt.legend()
plt.show()
```

```python
# Оценка λ = np.mean(x). Эта оценка представляет собой просто среднее значение выборки. 
#Однако среднее значение выборки не является хорошей оценкой для параметра 
#λ экспоненциального распределения, так как истинное значение 
#λ определяется как обратная величина среднего значения выборки.

import numpy as np
import matplotlib.pyplot as plt

# Функции для оценки параметра λ
def estimate_lambda_mean(x):
    return np.mean(x)

# Истинное значение параметра λ
true_lambda = 7

# Объемы выборок
sample_sizes = [10, 50, 100, 500, 1000]

# Число повторов для каждой выборки
num_repeats = 1000

# Для хранения оценок
estimates_mean = {size: [] for size in sample_sizes}

# Генерация выборок и оценка параметра λ
np.random.seed(42)
for size in sample_sizes:
    for _ in range(num_repeats):
        samples = np.random.exponential(1/true_lambda, size)
        estimates_mean[size].append(estimate_lambda_mean(samples))

# Построение графиков
plt.figure(figsize=(14, 8))
for size in sample_sizes:
    plt.plot(estimates_mean[size], label=f'Mean, Размер выборки = {size}')

plt.axhline(y=true_lambda, color='r', linestyle='--', label='True λ')
plt.xlabel('Испытания')
plt.ylabel('Оценка λ')
plt.title('Оценка λ для разного объема выборки')
plt.legend()
plt.show()
```

```python
# Оценка λ = np.log(2) / np.median(x). Эта оценка будет колебаться вокруг истинного значения 
# λ, особенно для больших выборок. Она устойчива к выбросам, но может быть менее точной, чем оценка максимального правдоподобия.

import numpy as np
import matplotlib.pyplot as plt

# Функции для оценки параметра λ

def estimate_lambda_median(x):
    return np.log(2) / np.median(x)

# Истинное значение параметра λ
true_lambda = 7

# Объемы выборок
sample_sizes = [10, 50, 100, 500, 1000]

# Число повторов для каждой выборки
num_repeats = 1000

# Для хранения оценок
estimates_mean = {size: [] for size in sample_sizes}

# Генерация выборок и оценка параметра λ
np.random.seed(42)
for size in sample_sizes:
    for _ in range(num_repeats):
        samples = np.random.exponential(1/true_lambda, size)
        estimates_mean[size].append(estimate_lambda_median(samples))

# Построение графиков
plt.figure(figsize=(14, 8))
for size in sample_sizes:
    plt.plot(estimates_mean[size], label=f'Mean, Размер выборки = {size}')

plt.axhline(y=true_lambda, color='r', linestyle='--', label='True λ')
plt.xlabel('Испытания')
plt.ylabel('Оценка λ')
plt.title('Оценка λ для разного объема выборки')
plt.legend()
plt.show()
```

```python
# Оценка λ = 1 / np.mean(x). Эта оценка является стандартной для экспоненциального распределения и должна точно сходиться к истинному значению 
# λ с увеличением объема выборки.

import numpy as np
import matplotlib.pyplot as plt

# Функции для оценки параметра λ

def estimate_lambda_mle(x):
    return 1 / np.mean(x)

# Истинное значение параметра λ
true_lambda = 7

# Объемы выборок
sample_sizes = [10, 50, 100, 500, 1000]

# Число повторов для каждой выборки
num_repeats = 1000

# Для хранения оценок
estimates_mean = {size: [] for size in sample_sizes}

# Генерация выборок и оценка параметра λ
np.random.seed(42)
for size in sample_sizes:
    for _ in range(num_repeats):
        samples = np.random.exponential(1/true_lambda, size)
        estimates_mean[size].append(estimate_lambda_mle(samples))

# Построение графиков
plt.figure(figsize=(14, 8))
for size in sample_sizes:
    plt.plot(estimates_mean[size], label=f'Mean, Размер выборки = {size}')

plt.axhline(y=true_lambda, color='r', linestyle='--', label='True λ')
plt.xlabel('Испытания')
plt.ylabel('Оценка λ')
plt.title('Оценка λ для разного объема выборки')
plt.legend()
plt.show()
```

```python
# Оценка λ = 1/(np.log(n)/n + np.mean(x)). Эта оценка будет иметь свойства, близкие к оценке максимального правдоподобия при больших 
# n, но может быть смещенной для малых выборок. Она стремится к 
# λ при увеличении объема выборки.

import numpy as np
import matplotlib.pyplot as plt

# Функции для оценки параметра λ

def estimate_lambda_modified(x, n):
    return 1 / (np.log(n) / n + np.mean(x))

# Истинное значение параметра λ
true_lambda = 7

# Объемы выборок
sample_sizes = [10, 50, 100, 500, 1000]

# Число повторов для каждой выборки
num_repeats = 1000

# Для хранения оценок
estimates_mean = {size: [] for size in sample_sizes}

# Генерация выборок и оценка параметра λ
np.random.seed(42)
for size in sample_sizes:
    for _ in range(num_repeats):
        samples = np.random.exponential(1/true_lambda, size)
        estimates_mean[size].append(estimate_lambda_modified(samples,size))

# Построение графиков
plt.figure(figsize=(14, 8))
for size in sample_sizes:
    plt.plot(estimates_mean[size], label=f'Mean, Размер выборки = {size}')

plt.axhline(y=true_lambda, color='r', linestyle='--', label='True λ')
plt.xlabel('Испытания')
plt.ylabel('Оценка λ')
plt.title('Оценка λ для разного объема выборки')
plt.legend()
plt.show()
```
