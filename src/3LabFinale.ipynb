```python
import numpy as np
import matplotlib.pyplot as plt

#        
def estimate_lambda(samples):
    return 1 / np.mean(samples)

#    
true_lambda = 7

#  
sample_sizes = [10, 50, 100, 500, 1000]

#     
num_repeats = 1000
    
#        
estimates = {size: [] for size in sample_sizes}

#             
np.random.seed(42)
for size in sample_sizes:
    for _ in range(num_repeats):
        samples = np.random.exponential(1/true_lambda, size)
        estimates[size].append(estimate_lambda(samples))

#  
plt.figure(figsize=(14, 8))
for size in sample_sizes:   
    plt.plot(estimates[size], label=f'  = {size}')

plt.axhline(y=true_lambda, color='r', linestyle='--', label='  ')
plt.xlabel('')
plt.ylabel(' ')
plt.title('     ')
plt.legend()
plt.show()
```

```python
#   = 1/(np.log(n)/n + np.mean(x)).     ,        
# n,       .    
#     .

import numpy as np
import matplotlib.pyplot as plt

#     

def estimate_lambda_modified(x, n):
    return 1 / (np.log(n) / n + np.mean(x))

#    
true_lambda = 7

#  
sample_sizes = [10, 50, 100, 500, 1000]

#     
num_repeats = 1000

#   
estimates_mean = {size: [] for size in sample_sizes}

#      
np.random.seed(42)
for size in sample_sizes:
    for _ in range(num_repeats):
        samples = np.random.exponential(1/true_lambda, size)
        estimates_mean[size].append(estimate_lambda_modified(samples,size))

#  
plt.figure(figsize=(14, 8))
for size in sample_sizes:
    plt.plot(estimates_mean[size], label=f'Mean,   = {size}')

plt.axhline(y=true_lambda, color='r', linestyle='--', label='True ')
plt.xlabel('')
plt.ylabel(' ')
plt.title('     ')
plt.legend()
plt.show()
```
